```python
import cv2
import numpy as np
from skimage.transform import resize

def video_mamonreader(cv2, filename):
    frames = np.zeros((30, 160, 160, 3), dtype=np.float64)
    i = 0
    print(frames.shape)
    vc = cv2.VideoCapture(filename)
    
    if not vc.isOpened():
        print("Error: Could not open video.")
        return None

    print("reading video")
    while i < 30:
        rval, frame = vc.read()
        if not rval:
            break
        frm = resize(frame, (160, 160, 3))
        if np.max(frm) > 1:
            frm = frm / 255.0
        frames[i] = frm
        i += 1
    
    vc.release()
    return frames[:i]  # return only the frames read

# Example usage
import time
millis = int(round(time.time() * 1000))
print("started at", millis)
vid = video_mamonreader(cv2, 'dvpalfight.dvd')
millis2 = int(round(time.time() * 1000))
print("time processing", millis2 - millis)

if vid is not None:
    print("Successfully read video frames.")
else:
    print("Failed to read video frames.")
```
